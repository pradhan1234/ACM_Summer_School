```python
#Task 1: Read each GOT file, get time stamps and non dialogues and write them to result file

import os
import re

fileList = !ls 'Demo/subtitles/'
nonDlgFile = open('NonDlgFile.txt', 'a')
timestampFile = open('TimeStampFile.txt', 'a')

for subtitleFile in fileList:
    f = open('Demo/subtitles/'+subtitleFile, 'r')
    content = f.read()
    
    nonDlgPtn = re.compile('\(.*\)')
    nonDlgList = nonDlgPtn.findall(content)
    
    for nonDlg in nonDlgList:
        nonDlgFile.write(nonDlg+"\n")
    nonDlgFile.write("--------------------------------------\n")

    timestampPtn = re.compile('\d\d:\d\d:\d\d,\d\d\d\s*-->\s*\d\d:\d\d:\d\d,\d\d\d')
    timestampList = timestampPtn.findall(content)
    
    for timestamp in timestampList:
        timestampFile.write(timestamp+"\n")
        
    
    
f.close()
nonDlgFile.close()
timestampFile.close()
```
