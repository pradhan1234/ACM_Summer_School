```python
refinedSubs = list()
for thing in subs:
    epi = list()
    for item in thing.lower().split('\r\n\r\n'):
        stuff = item.split('\r\n')
        setVal = 0
        refinedStuff = ''
        for entry in stuff[2:]:
            for chara in entry.replace('<i>','').replace('</i>',''):
                if chara == '(':
                    setVal = 1
                if setVal == 0:
                    if chara in '".,!\':;=?([])-':
                        refinedStuff += ' '
                    elif chara in string.punctuation:
                        refinedStuff += ''
                    else:
                        refinedStuff += chara
                if chara == ')':
                    setVal = 0
            if '<' in refinedStuff:
                refinedStuff = ''
                    
        if len(refinedStuff.strip().strip('\t')) > 1:
            epi.append(refinedStuff)
    refinedSubs.append(epi)
```
